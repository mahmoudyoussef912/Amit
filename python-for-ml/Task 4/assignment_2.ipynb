```python
import os
import random
def thanos_project(folder_name,num_of_file):
    """
    this function will do this :
    1 open folder
    2 creat number of files inside it 
    3 check the number of files in that folder
    4 delete random half of this folder
    """
    # 1 open folder
    if not os.path.exists(folder_name):
        os.mkdir(folder_name)
    os.chdir(folder_name)
    
    #2 creat number of files inside it
    for i in range(num_of_file):
        with open(f"file_{i}.txt", "w") as f:
            f.write("hello worled/n")
            
    #3 check the number of files in that folder    
    files = os.listdir()
    print("number of files : " , len(files))
    
    #4 delete random half of this folder
    half = len(files) // 2
    file_to_del = random.sample(files,half)
    for file in file_to_del :
        os.remove(file)
    
    # check remaining files
    remaining_files = os.listdir()
    print("Files after delete:", len(remaining_files))

# test thanos_project function
thanos_project('test',20)
```
